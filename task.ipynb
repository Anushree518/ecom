```python
import pandas as pd
```

```python
df1=pd.read_csv('tab_data1.csv')
```

```python
df1
```

```python
df2=pd.read_csv('tab_data2.csv')
```

```python
df2
```

```python
df = pd.merge(df1, df2, left_on='User ID', right_on='uid')
```

```python
df
```

```python
# calculate the total number of statements and reasons
df['total'] = df['total_statements'] + df['total_reasons']
```

```python
# sort the dataframe by Total, then by Name (alphabetically)
df=df.sort_values(['total','Name'], ascending=[False, True])
```

```python
# create a ranking column based on the sorted dataframe, with preference to alphabetical order in case of tie
df['Rank']= df[['total','Name']].rank(method='min', ascending=[False, True]).iloc[:, 0].astype(int)
```

```python
# select only the columns needed for the leaderboard
leaderboard= df[['Rank','Name','uid','total_statements','total_reasons']]
leaderboard.columns=['Rank','Name','uid','No. of Statements', 'No. of Reasons']
leaderboard=leaderboard.sort_values(by=['Rank'])
```

```python
# display the leaderboard
print(leaderboard.to_string(index=False))
```

```python
# group by Team Name and calculate average statements and reasons
team_stats= df.groupby('Team Name').agg({'total_statements':'mean', 'total_reasons': 'mean'})
```

```python
# add a new column for team rank
team_stats['Team Rank']=team_stats['total_statements'].rank(ascending=False).astype(int)
```

```python
# sort the dataframe based on 'Team Rank' column in descending order
team_stats=team_stats.sort_values(by=['Team Rank'], ascending=False)
```

```python
# rename columns and reset index
team_stats=team_stats.rename(columns={'total_statements': 'Average Statements','total_reasons': 'Average Reasons'})
team_stats=team_stats.reset_index()
```

```python
# reorder columns and format decimals
team_stats=team_stats[['Team Rank','Team Name','Average Statements', 'Average Reasons']]
team_stats[['Average Statements','Average Reasons']]= team_stats[['Average Statements', 'Average Reasons']].round(2)
team_stats=team_stats.sort_values(by=['Team Rank'])
```

```python
# print the final table
print(team_stats.to_string(index=False))
```
